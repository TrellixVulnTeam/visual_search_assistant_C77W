## Create Splits

```python
import shutil, os

src_root = '/home/ubuntu/visual_search_assistant/aligned_images_DB'
dst_root = '/home/ubuntu/visual_search_assistant/youtube_split_'

train_dir = os.path.join(dst_root,'train')
test_dir = os.path.join(dst_root,'test')

if not os.path.exists(dst_root):
    os.makedirs(train_dir)
    os.makedirs(test_dir)
split_ratio = 0.8

for identity in sorted(os.listdir(src_root)):
    identity_dir = os.path.join(src_root,identity)
    
    videos = sorted(os.listdir(os.path.join(src_root,identity)))
    num_train = int(len(videos)*split_ratio)
#     import pdb;pdb.set_trace()
    if num_train==0 or num_train == len(videos):
        # skip import
        print('Not enough videos for %s,,,,%d %d,,skipping' % (identity,num_train,len(videos)))
        continue
    
    # train split
    for video in videos[:num_train]:
        src = os.path.join(identity_dir,video)
        dst = os.path.join(train_dir,identity)
        if not os.path.exists(dst):
            os.mkdir(dst)
        for file in os.listdir(src):
            shutil.copy(os.path.join(src,file),dst)
#             import pdb;pdb.set_trace()
    # test split
    for video in videos[num_train:]:
        src = os.path.join(identity_dir,video)
        dst = os.path.join(test_dir,identity)
        if not os.path.exists(dst):
            os.mkdir(dst)
        for file in os.listdir(src):
            shutil.copy(os.path.join(src,file),dst)
    
    print('Done processing %s' % identity)
        
```

## Check split statistics

```python
print("Num of identities in train dir: %d" %len(os.listdir(train_dir)))
print("Num of identities in test dir: %d"%len(os.listdir(test_dir)))

assert len(os.listdir(train_dir)) == len(os.listdir(test_dir))
```
